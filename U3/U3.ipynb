## Aufgabe 1

Schreiben Sie ein MATLAB-Programm, das die DGL
$$\frac{dx}{dt}-2\cdot \sin^3(x) - 1 = 0$$
mit der Anfangsbedingung $x(0) = 1$ numerisch löst und die Lösungsfunktion grafisch im Intervall $0 \leq t \leq 5$ darstellt.

**Hinweis:** Vergessen Sie nicht, die DGL in einem gesonderten function-file zu definieren.

```matlab
clear all;
clf;
tspan = [0 5];
x0 = 1;
% defining ODE using function handle
f = @(t,x)(2*sin(x)^3+1);
[t, x] = ode45(f, tspan, x0);

figure(1)
plot(t,x,'-o')

set(gcf,'Position', [0 0 1280 720])
grid on
title('A1: $\dot{x} = 2sin^3(x)+1$', 'Interpreter', 'latex')
xlabel('t')
ylabel('x')
hold off
```

## Aufgabe 2

Schreiben Sie ein MATLAB-Programm, das die DGL eines ungedämpften Federpendels numerisch löst:
$$\frac{d^2}{dt^2}x + \frac{D}{m}x = 0$$
Die Federkonstante beträgt $D=196\,N/m$ und die schwingende Masse sei $m=2\,kg$. Wählen Sie selbst eine Anfangsbedingung. Stellen Sie die Lösungsfunktion grafisch in einem sinnvollen Intervall dar.

**Hinweis:** Vergessen Sie nicht, die DGL in einem gesonderten function-file zu definieren.

```matlab
clear all;
clf;
D = 196; % Newton per metre
m = 2; % kg
w0 = sqrt(D/m);
disp(sprintf('Resonant frequency: %.3f s^-1', w0))

tspan = [0, 1];
v0 = 7;
x0 = 15;

f = @(t,u)[u(2); -(D/m)*u(1)];

[t,x] = ode45(f, tspan, [x0 v0]);

figure(2)
yyaxis left
plot(t,x(:,1))
ylabel('Elongation')

yyaxis right
plot(t,x(:,2))
ylabel('Velocity');

set(gcf,'Position', [0 0 1280 720])
title('A2: $\ddot{x}+\frac{D}{m}x = 0$', 'Interpreter', 'latex');
xlabel('Time / t');
legend('Velocity', 'Elongation');
grid on
hold off

%function dudt = odeA2(t,u)
    % u(2) := velocity
    % u(1) := elongation
%    dudt = [u(2); -(D/m)*u(1)];
%end
```
